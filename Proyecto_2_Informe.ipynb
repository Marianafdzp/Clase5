# **Resumen**


Las Interfaces Cerebro-Computador (BCI) requieren la decodificación precisa de estados mentales a partir de señales de electroencefalografía (EEG). Mientras que el análisis tradicional se centra en la Densidad Espectral de Potencia (PSD), este proyecto evaluó la viabilidad de mejorar la clasificación de estados BCI implementando algoritmos de Machine Learning alimentados por características multidominio. Utilizando los datos preprocesados del Proyecto 1, se extrajeron tres índices avanzados: la Entropía Aproximada (para evaluar dinámicas no lineales), la Coherencia (para conectividad funcional) y Patrones Espaciales Comunes o CSP (para discriminación topográfica).

# **Introducción**


El desarrollo de sistemas BCI orientados a la asistencia o el control requiere algoritmos capaces de traducir con alta fidelidad la actividad cerebral. En el **Proyecto 1**, la caracterización de los estados de interés se realizó mediante la Densidad Espectral de Potencia (PSD) para evidenciar fenómenos fisiológicos como la desincronización (ERD) en ritmos específicos. Si bien este enfoque clásico permitió observar "qué frecuencias estaban presentes", asume erróneamente que la señal EEG es lineal y estacionaria, limitando la capacidad de los modelos para distinguir comandos frente a la alta variabilidad inter-sujeto e intra-sujeto.

Las bioseñales cerebrales provienen de un sistema dinámico y altamente interconectado. Por lo tanto, el presente proyecto, surge de la necesidad de superar las limitaciones del análisis puramente espectral. El objetivo es diseñar y evaluar un sistema de clasificación automatizado mediante Machine Learning, alimentando los modelos con una extracción de características multidominio.

En lugar de depender de una sola perspectiva, se integran tres índices neurofisiológicos complementarios:

* La Entropía Aproximada (ApEn), que cuantifica la predictibilidad y el caos temporal de la señal.
* La Coherencia, que evalúa cómo se comunican y sincronizan las diferentes regiones corticales (conectividad funcional).
* Los Patrones Espaciales Comunes (CSP), que proyectan la señal para maximizar la varianza espacial entre las distintas condiciones del usuario.

La integración de estas métricas no lineales, espaciales y de red busca robustecer el vector de características, disminuyendo el ruido y permitiendo que los algoritmos de aprendizaje automático alcancen precisiones óptimas para un sistema BCI en la vida real.

### Justificación y Descripción de la Selección de Índices

Con base en el análisis del estado del arte evidenciado en la tabla anterior, y comprendiendo las limitaciones del **Proyecto 1** (donde la Densidad Espectral de Potencia - PSD solo nos permitió observar fenómenos energéticos promedio de forma lineal), se justifica la selección de la **Entropía Aproximada (ApEn)**, la **Coherencia** y los **Patrones Espaciales Comunes (CSP)** como la estrategia más prometedora para este segundo ciclo del proyecto.

La idoneidad de estos tres índices radica en los siguientes factores técnicos:

* **Complementariedad Ortogonal de Dominios:** Uno de los mayores retos en *Machine Learning* es evitar la redundancia en los datos de entrada (*Feature Vector*). Esta selección ataca el problema desde tres dominios totalmente diferentes, los cuales se describen a continuación:
    * **Dominio Espacial (Topográfico) - Patrones Espaciales Comunes (CSP):** El CSP [3] es un algoritmo matemático de filtrado espacial supervisado que utiliza las matrices de covarianza del EEG. Su función es calcular proyecciones lineales que maximizan la varianza de la señal para una clase (ej. movimiento derecho) mientras la minimizan para otra (ej. movimiento izquierdo). En este modelo, el CSP actúa como una lupa espacial orientada por clases, encontrando exactamente qué regiones de la corteza cerebral se activan de manera opuesta durante las diferentes condiciones.
    * **Dominio de Red (Conectividad) - Coherencia:** La Coherencia [2] es una métrica espectral que cuantifica la correlación cruzada y la sincronización de fase entre dos bioseñales en función de la frecuencia. A nivel fisiológico, evalúa la sincronización funcional; por lo tanto, no asume el cerebro como un grupo de sensores aislados, sino que mide matemáticamente cómo "dialogan" y se acoplan diferentes áreas (ej. el lóbulo frontal con la corteza motora) para ejecutar en conjunto la tarea cognitiva o motora.
    * **Dominio Temporal No-Lineal (Complejidad) - Entropía Aproximada (ApEn):** La ApEn [1] es un estadístico derivado de la teoría del caos que cuantifica la cantidad de regularidad y la impredecibilidad de las fluctuaciones en una serie de tiempo (valores bajos indican una señal predecible y valores altos indican caos). Su inclusión permite medir la tasa de generación de información temporal, detectando cambios sutiles y micro-transiciones no lineales que los métodos basados en promedios estáticos, como la PSD del Proyecto 1, borran por completo.

* **Sinergia directa con algoritmos de Machine Learning:** Los clasificadores solicitados para este proyecto, como la Máquina de Vectores de Soporte (**SVM**) o el *Extreme Gradient Boosting* (**XGBoost**), alcanzan su máximo potencial cuando se enfrentan a fronteras de decisión geométricamente ricas. Al fusionar la linealidad topográfica del CSP con la profundidad dinámica y de conectividad de la Entropía y la Coherencia, los algoritmos logran converger más rápido, evitando el sobreajuste (*overfitting*) a pesar del ruido biológico.

* **Evolución del Baseline de Proyecto 1:** Mientras que en el Proyecto 1 nos limitamos a corroborar fisiológicamente la existencia de la Desincronización Relacionada a Eventos (ERD/ERS), la extracción simultánea de ApEn, Coherencia y CSP convierte nuestro sistema analítico en una verdadera herramienta de **predicción de estados**. Esto transforma el análisis pasivo de los datos en un motor de BCI factible y trasladable a necesidades reales de la ingeniería de rehabilitación.

# **Perspectivas y Proyectos Futuros en BCI impulsados por Machine Learning**

El análisis de la literatura reciente (2016-2026) demuestra que la transición hacia índices multidominio es fundamental, pero su verdadero potencial solo se alcanza al integrarlos con algoritmos predictivos. A partir de los métodos evaluados en el estado del arte, los futuros proyectos de investigación clínica y tecnológica apuntan a evolucionar las Interfaces Cerebro-Computador (BCI) utilizando el Machine Learning como eje central:

**Integración de Conectividad Funcional con Machine Learning Basado en Grafos**
El futuro de la decodificación cognitiva apunta a dejar de tratar los canales de EEG como sensores aislados para modelarlos como redes biológicas complejas. Como lo demostraron Zhang *et al.* [2], los próximos proyectos buscan alimentar arquitecturas de *Machine Learning* avanzado (específicamente Redes Neuronales de Grafos - GCN) directamente con matrices de Coherencia o de Valor de Bloqueo de Fase (PLV) [8]. A diferencia de los clasificadores estándar, estos modelos de ML aprenderán de forma automática la topología de la red cerebral, identificando patrones de comunicación entre hemisferios que revelan intenciones motoras mucho antes de que se ejecuten.

**Modelos de Machine Learning de "Cero Calibración" (Transfer Learning)**
Uno de los mayores cuellos de botella del ML clásico en BCI es que un modelo (como un SVM o XGBoost) entrenado para un paciente rara vez funciona para otro debido a la variabilidad cerebral. Basándose en enfoques geométricos como los de Zanini *et al.* [5], los proyectos futuros se centrarán en desarrollar modelos de *Machine Learning* con capacidades de *Transfer Learning* (Aprendizaje Transferido). Al proyectar las características topográficas (CSP [3]) en espacios de Geometría Riemanniana, se busca crear algoritmos universales pre-entrenados que puedan adaptarse a un usuario completamente nuevo en segundos, eliminando las tediosas sesiones de calibración clínica.

**Machine Learning en el Borde (Edge ML / TinyML) para Tiempo Real**
Para que los sistemas BCI sean viables en la vida diaria del paciente, los algoritmos predictivos deben salir de las grandes computadoras de laboratorio. Trabajos con preprocesamiento adaptativo, como el uso de *Sparse CSP* [4], abren la puerta al desarrollo de *TinyML* (Machine Learning en microcontroladores). El objetivo futuro es incorporar clasificadores ligeros (como K-Nearest Neighbors o pequeños ensambles de árboles) directamente en diademas portátiles (*wearables*). Estos dispositivos calcularán métricas rápidas como la Entropía Aproximada [10] y ejecutarán inferencias de ML en cuestión de milisegundos para monitorear fatiga, estrés o comandos de sillas de ruedas en tiempo real.

**Ingeniería de Características Híbrida para Clasificadores de Ensamble**
Para llevar la precisión de los sistemas no invasivos a estándares comerciales (superiores al 95%), los proyectos venideros continuarán optimizando la fase de extracción de características (*Feature Engineering*) para modelos de Machine Learning tradicionales. Tal como sugieren investigaciones recientes [1], [9], el futuro radica en entrenar algoritmos de ensamble potentes (como *Extreme Gradient Boosting* - XGBoost o *Random Forest*) suministrándoles vectores híbridos. Esto implica fusionar en una misma matriz de entrada el comportamiento caótico del cerebro (Entropía Muestral) con análisis precisos de tiempo-frecuencia (Wavelet), dándole al modelo de ML fronteras de decisión mucho más ricas y robustas frente al ruido biológico.

# **Plan de análisis**

Para cumplir con los objetivos del proyecto y superar el rendimiento de la clasificación basada únicamente en la Densidad Espectral de Potencia (PSD) del Proyecto 1, se propone el siguiente flujo de trabajo analítico. Este plan integra la extracción de características multidominio con algoritmos avanzados de *Machine Learning* y *Deep Learning*.

El análisis se dividirá en cinco fases fundamentales:

### Fase 1: Preprocesamiento y Segmentación de Datos (Epoching)
Se procedió a la lectura y limpieza de las señales continuas provenientes de las tareas de imaginación motora (runs 4, 8 y 12 del paradigma experimental).
1. **Selección de Canales:** Se aislarán exclusivamente los canales **C3** y **C4**, ubicados sobre la corteza motora primaria, responsables de la actividad sensoriomotora de las extremidades superiores.
2. **Filtrado Espacial y Frecuencial:** Se aplicará un filtro pasa-banda FIR (ventana de Hamming) entre 8 Hz y 30 Hz para eliminar el ruido de baja frecuencia y los artefactos musculares de alta frecuencia, aislando los ritmos de interés.
3. **Segmentación (Epoching):** La señal continua se dividió en ventanas temporales (épocas) de 4 segundos (tmin=0, tmax=4), categorizadas bajo tres etiquetas numéricas directas para el algoritmo de clasificación:
   * **0:** Estado de reposo (TASK2T0).
   * **1:** Imaginación motora de la mano izquierda (TASK2T1).
   * **2:** Imaginación motora de la mano derecha (TASK2T2).

### Fase 2: Extracción de Características Multidominio (Feature Engineering)
Por cada época segmentada, se extrajeron características matemáticas divididas explícitamente en dos bandas rítmicas fundamentales: **Banda Mu (8-12 Hz)** y **Banda Beta (13-30 Hz)**. Se calcularon cuatro índices diferentes, generando un vector de características robusto:

1. **Dominio Frecuencial (Densidad Espectral de Potencia - PSD):** Mediante el método de Welch (1 a 40 Hz), se calculó la potencia total y se extrajo la potencia relativa de las bandas Mu y Beta para los canales C3 y C4.
2. **Dominio No Lineal (Entropía Aproximada - ApEn):** Se aplicó el algoritmo de Entropía Aproximada (orden = 2) sobre las señales filtradas de C3 y C4, evaluando el nivel de complejidad y predictibilidad temporal de la señal durante la imaginación motora.
3. **Dominio de Conectividad (Coherencia):** Se calculó el promedio de la coherencia cruzada espectral entre los canales C3 y C4, cuantificando el grado de sincronización de fase entre el hemisferio derecho e izquierdo durante las tareas.
4. **Dominio Espacial (Patrones Espaciales Comunes - CSP):** Se ajustó un filtro espacial supervisado (entrenado exclusivamente con las épocas de imaginación izquierda y derecha) para maximizar la separabilidad de la varianza. De este modelo se extrajeron 2 componentes (log-varianza) por cada banda de frecuencia.


### Fase 3: Análisis estadístico
Para evaluar rigurosamente el poder discriminativo de cada índice extraído antes de la clasificación, se aplicará la siguiente ruta de pruebas estadísticas:

1. Verificación de normalidad: Se aplicará la prueba de D'Agostino-Pearson a cada índice extraído, evaluando su distribución de forma independiente en las tres condiciones (reposo, imaginación mano derecha, imaginación mano izquierda).

2. Prueba de hipótesis general: Para comprobar si existen diferencias significativas entre las tres condiciones, la elección de la prueba dependerá de la normalidad:

- Si los datos en las tres condiciones presentan normalidad: Se utilizará un ANOVA de una vía.

- Si los datos no presentan normalidad (al menos en una condición): Se utilizará la prueba no paramétrica de Kruskal-Wallis.

3. Comparaciones post-hoc: Solo en caso de que la prueba de hipótesis anterior resulte estadísticamente significativa (valor p menor a 0.05), se procederá a identificar entre qué pares exactos de condiciones existen diferencias:

- Si se utilizó ANOVA: Se aplicará la prueba de Tukey HSD.

- Si se utilizó Kruskal-Wallis: Se aplicará la prueba de Dunn con corrección de Bonferroni (para controlar la tasa de error por comparaciones múltiples).

### Fase 4: Entrenamiento y Clasificación con Machine Learning
La matriz de características alimentará diferentes arquitecturas de clasificación predictiva para evaluar cuál se adapta mejor a la topología de los datos:
1.  **Máquinas de Vectores de Soporte (SVM):** Se implementará un modelo SVM (utilizando kernels lineales o RBF) para encontrar el hiperplano óptimo que separe las clases maximizando el margen, ideal para los datos proyectados por CSP.
2.  **Extreme Gradient Boosting (XGBoost):** Se utilizará este potente algoritmo de ensamble basado en árboles de decisión, el cual es altamente eficiente para manejar características de múltiples dominios (como la mezcla de Coherencia y Entropía) y mitigar el sobreajuste.
3.  **Arquitecturas de Redes Neuronales:** Se diseñarán, entrenarán y discutirán por lo menos tres arquitecturas diferentes de redes neuronales (ej. Perceptrón Multicapa - MLP de distintas profundidades) para evaluar la capacidad del *Deep Learning* en la clasificación de este conjunto de características.

### Fase 5: Validación y Evaluación de Rendimiento
Para asegurar que los modelos sean generalizables y no presenten sobreajuste (*overfitting*), se aplicará el siguiente esquema de evaluación:
1.  **Validación Cruzada (*k-fold Cross-Validation*):** Todo el proceso de entrenamiento y evaluación se realizará utilizando un esquema de *k-fold* (ej. $k=5$ o $k=10$), dividiendo iterativamente los datos en conjuntos de entrenamiento y prueba para asegurar robustez estadística.
2.  **Métricas de Desempeño:** La capacidad de discriminación de los algoritmos se medirá mediante la **Exactitud (*Accuracy*)**, **Sensibilidad** y **Especificidad**.
3.  **Análisis de Errores:** Se generarán e interpretarán las Matrices de Confusión para cada modelo, evaluando si existen falsos positivos o falsos negativos sesgados hacia alguna clase en particular.
4.  **Comparación Final:** Se contrastarán los resultados obtenidos en este pipeline de *Machine Learning* multidominio contra el rendimiento base (baseline) observado fisiológicamente en el Proyecto 1.
